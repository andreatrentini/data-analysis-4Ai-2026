```python
import math

def isPrimo(numero):
    if numero <= 1:
        return False
    else:
        for divisore in range(2, int(math.sqrt(numero)) + 1):
            if numero % divisore == 0:
                return False
        return True

numero = 2
nrPrimi = 0
while nrPrimi < 100:
    if isPrimo(numero):
        print(numero, end=", ")
        nrPrimi += 1
    numero += 1
```
