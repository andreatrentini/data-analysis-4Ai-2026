```python
bravi_mate = dfDaCSV[dfDaCSV['matematica'] >= 8]
bravi_mate
bravi_stem = dfDaCSV[(dfDaCSV['matematica'] >= 8) & (dfDaCSV['informatica'] >= 8)]
bravi_stem

studenti_4A = dfDaCSV[dfDaCSV['classe'] == '4A']
studenti_4A

studenti_altre_classi = dfDaCSV[dfDaCSV['classe'].isin(['4B', '4C'])]
studenti_altre_classi

studenti_altre_classi_2 = dfDaCSV[~dfDaCSV['classe'].isin(['4A'])]
studenti_altre_classi_2
```

```python
lista_materie = ['matematica','italiano','informatica','fisica','storia']
dfDaCSV['media'] = dfDaCSV[lista_materie].mean(axis=1).round(2)

dfDaCSV['media_scien'] = dfDaCSV[['matematica','informatica','fisica']].mean(axis=1).round(2)
dfDaCSV['media_uma'] = dfDaCSV[['italiano','storia']].mean(axis=1).round(2)

dfDaCSV
```

```python
nuovo_dfDaCSV_rinominato = dfDaCSV.rename(columns={
    'media': 'media_totale'
})

nuovo_dfDaCSV_rinominato
```

```python
df_ordinato = nuovo_dfDaCSV_rinominato.sort_values('media_totale', ascending=False)
df_ordinato
```

```python
dfDaCSV['totale'] = dfDaCSV[['matematica', 'italiano', 'informatica', 'fisica', 'storia']].sum(axis=1)
dfDaCSV.drop(columns=['totale'], inplace=True)
dfDaCSV.dtypes

dfDaCSV['classe'].value_counts()
dfDaCSV['classe'].value_counts(normalize=True).round(2) * 100

dfDaCSV['sesso'].unique()
dfDaCSV['sesso'].nunique()
```
