Python distingue maiuscole e minuscole: `nome`, `Nome` e `NOME` sono tre variabili diverse.

## 5. Variabili e Tipi di Dati

Python usa la **tipizzazione dinamica**: il tipo viene dedotto automaticamente dal valore assegnato.

### Tipi built-in fondamentali
| Tipo | Descrizione | Esempio |
|------|-------------|--------|
| `int` | Intero (precisione arbitraria) | `42`, `-7`, `0` |
| `float` | Virgola mobile (IEEE 754) | `3.14`, `-0.5`, `1e10` |
| `str` | Stringa di testo (Unicode) | `"ciao"`, `'mondo'` |
| `bool` | Booleano | `True`, `False` |
| `NoneType` | Assenza di valore | `None` |
| `complex` | Numero complesso | `2+3j` |

### Type casting
Conversione esplicita tra tipi con `int()`, `float()`, `str()`, `bool()`.


````python
# Dichiarazione e assegnazione
eta = 17                    # int
altezza = 1.75              # float
nome = 'Alice'              # str
promosso = True             # bool
voto_extra = None           # NoneType
numero_complesso = 2 + 3j   # complex

# type() restituisce il tipo
print(type(eta))            # <class 'int'>
print(type(altezza))        # <class 'float'>
print(type(nome))           # <class 'str'>
print(type(promosso))       # <class 'bool'>
print(type(voto_extra))     # <class 'NoneType'>

# Assegnazione multipla
x = y = z = 0
a, b, c = 1, 2, 3
print(f'a={a}, b={b}, c={c}')

# Type casting
stringa_num = '42'
numero = int(stringa_num)           # str → int
print(f'"{stringa_num}" → {numero}, tipo: {type(numero).__name__}')

pi_str = str(3.14159)               # float → str
print(f'3.14159 → "{pi_str}"')

print(bool(0), bool(1), bool(''), bool('ciao'))  # 0/''→False, altri→True

# int con basi diverse
binario = 0b1010    # base 2  → 10
ottale = 0o17       # base 8  → 15
esadec = 0xFF       # base 16 → 255
print(f'Binario: {binario}, Ottale: {ottale}, Esadecimale: {esadec}')
````

| Operatore | Descrizione |
|-----------|-------------|

`+=`, `-=`, `*=`, `/=`, `//=`, `%=`, `**=`


````python
# Operatori aritmetici
a, b = 17, 5
print(f'{a} + {b}  = {a + b}')
print(f'{a} - {b}  = {a - b}')
print(f'{a} * {b}  = {a * b}')
print(f'{a} / {b}  = {a / b}')
print(f'{a} // {b} = {a // b}')
print(f'{a} % {b}  = {a % b}')
print(f'{a} ** {b} = {a ** b}')

# Operatori di confronto
print(f'\n--- Confronto ---')
print(f'10 == 10: {10 == 10}')
print(f'10 != 5:  {10 != 5}')
print(f'3 > 7:    {3 > 7}')
print(f'5 <= 5:   {5 <= 5}')

# Operatori logici
print(f'\n--- Logici ---')
print(f'True and False: {True and False}')
print(f'True or False:  {True or False}')
print(f'not True:       {not True}')

# Assegnazione composta
n = 10
n += 5;  print(f'Dopo +=5: {n}')
n *= 2;  print(f'Dopo *=2: {n}')
n //= 3; print(f'Dopo //=3: {n}')

# Operatori speciali: is, in
lista = [1, 2, 3]
print(f'3 in lista:  {3 in lista}')
print(f'5 not in lista: {5 not in lista}')
x = None
print(f'x is None: {x is None}')
````

## 7. Input e Output

### `print()`
Accetta argomenti multipli, supporta `sep` e `end` per personalizzare l'output.

### `input()`
Legge una stringa dall'utente. **Restituisce sempre una stringa**: convertire se necessario.

### F-string (Python 3.6+)
Le f-string sono il metodo moderno e preferito per il formatting:
```python
nome = 'Alice'
f'Ciao, {nome}!'
```

### Modificatori di formato
| Formato | Effetto | Esempio |
|---------|---------|--------|
| `:.2f` | 2 decimali | `3.14` |
| `:10` | larghezza 10 | `'   Alice'` |
| `:<10` | allineato sx | `'Alice   '` |
| `:>10` | allineato dx | `'   Alice'` |
| `:^10` | centrato | `' Alice  '` |
| `:,` | separatore migliaia | `1,000,000` |
| `:b` | binario | `1010` |
| `:x` | esadecimale | `ff` |


````python
# print() con vari argomenti
print('uno', 'due', 'tre')                    # sep=' ' default
print('uno', 'due', 'tre', sep='-')           # sep personalizzato
print('riga 1', end=' | '); print('riga 2')   # end personalizzato

# F-string
nome = 'Mario'
eta = 17
media = 8.666666
print(f'Studente: {nome}, Età: {eta}')
print(f'Media voti: {media:.2f}')              # 2 decimali
print(f'Media voti: {media:.4f}')              # 4 decimali

# Allineamento con f-string
print(f'{"Nome":<15} {"Voto":>6}')
print(f'{"Alice":<15} {9.5:>6.1f}')
print(f'{"Bob":<15} {7.8:>6.1f}')
print(f'{"Carla":<15} {8.2:>6.1f}')

# Separatori e formati numerici
grande = 1_000_000
print(f'Numero grande: {grande:,}')
print(f'In binario: {255:b}')
print(f'In esadecimale: {255:#x}')

# Espressioni nelle f-string (Python 3.12: ancora più potenti)
valori = [3, 7, 2, 9, 1]
print(f'Max: {max(valori)}, Min: {min(valori)}, Somma: {sum(valori)}')
````

## 8. Selezione if/elif/else

La selezione condizionale permette di eseguire blocchi di codice differenti in base a condizioni.

### Struttura
```python
if condizione1:
    # blocco 1
elif condizione2:
    # blocco 2
else:
    # blocco default
```

### Operatore ternario
Espressione condizionale su una riga:
```python
risultato = valore_se_vero if condizione else valore_se_falso
```

### match-case (Python 3.10+)
Equivalente del `switch` di altri linguaggi, ma molto più potente grazie al **pattern matching strutturale**.


````python
# if / elif / else classico
voto = 7.5

if voto >= 9:
    giudizio = 'Ottimo'
elif voto >= 7:
    giudizio = 'Buono'
elif voto >= 6:
    giudizio = 'Sufficiente'
else:
    giudizio = 'Insufficiente'

print(f'Voto {voto}: {giudizio}')

# Condizioni composte
eta = 18
patente = True
if eta >= 18 and patente:
    print('Può guidare')
else:
    print('Non può guidare')

# Operatore ternario
x = 42
risultato = 'pari' if x % 2 == 0 else 'dispari'
print(f'{x} è {risultato}')

# Valori falsy: 0, '', [], {}, None, False
valori = [0, 1, '', 'ciao', [], [1], None, False, True]
for v in valori:
    print(f'  bool({repr(v)}) = {bool(v)}')
````

````python
# match-case (Python 3.10+) – pattern matching strutturale
def analizza_comando(comando):
    match comando.split():
        case ['quit'] | ['exit']:
            return 'Uscita dal programma'
        case ['help']:
            return 'Mostra aiuto'
        case ['vai', direzione] if direzione in ('nord', 'sud', 'est', 'ovest'):
            return f'Vai verso {direzione}'
        case ['vai', _]:
            return 'Direzione non valida'
        case []:
            return 'Comando vuoto'
        case _:
            return f'Comando sconosciuto: {comando}'

comandi = ['quit', 'help', 'vai nord', 'vai alto', 'ciao mondo', '']
for c in comandi:
    print(f'  [{c!r}] → {analizza_comando(c)}')

# match con strutture dati
def descrivi_punto(punto):
    match punto:
        case (0, 0):
            return 'Origine'
        case (x, 0):
            return f'Sull\'asse X in {x}'
        case (0, y):
            return f'Sull\'asse Y in {y}'
        case (x, y):
            return f'Punto ({x}, {y})'

print(descrivi_punto((0, 0)))
print(descrivi_punto((3, 0)))
print(descrivi_punto((0, -2)))
print(descrivi_punto((4, 5)))
````

## 9. Iterazione for/while

### Ciclo `for`
Itera su qualsiasi oggetto **iterabile** (lista, stringa, range, dizionario…).

### `range(start, stop, step)`
Genera una sequenza di interi: `range(5)` → 0,1,2,3,4

### Ciclo `while`
Esegue un blocco finché la condizione è `True`.

### Controllo del flusso nei cicli
| Istruzione | Effetto |
|------------|--------|
| `break` | Esce immediatamente dal ciclo |
| `continue` | Salta all'iterazione successiva |
| `else` | Eseguito se il ciclo termina senza `break` |


````python
# for su lista e stringa
frutti = ['mela', 'banana', 'ciliegia']
for frutto in frutti:
    print(f'  Frutto: {frutto}')

# range
print('\nRange 0-4:', list(range(5)))
print('Range 2-9 step 2:', list(range(2, 10, 2)))
print('Range inverso:', list(range(5, 0, -1)))

# enumerate() – indice + valore
print('\nCon enumerate:')
for i, frutto in enumerate(frutti, start=1):
    print(f'  {i}. {frutto}')

# zip() – itera su più sequenze contemporaneamente
nomi = ['Alice', 'Bob', 'Carla']
voti = [8.5, 7.0, 9.2]
print('\nCon zip:')
for nome, voto in zip(nomi, voti):
    print(f'  {nome}: {voto}')

# break e continue
print('\nbreak appena trovo il 5:')
for n in range(10):
    if n == 5:
        break
    print(n, end=' ')

print('\nSkip i numeri pari:')
for n in range(10):
    if n % 2 == 0:
        continue
    print(n, end=' ')

# else sul ciclo for
print('\n\nRicerca con for-else:')
dati = [3, 7, 2, 9, 1]
target = 4
for v in dati:
    if v == target:
        print(f'{target} trovato!')
        break
else:
    print(f'{target} non trovato nel ciclo')
````

````python
# while
n = 1
print('Potenze di 2:')
while n <= 512:
    print(n, end=' ')
    n *= 2

# while con input simulato
print('\n\nSimulazione indovina il numero:')
import random
random.seed(42)
segreto = random.randint(1, 20)
tentativi = [10, 15, segreto]  # simuliamo i tentativi
indice = 0

while indice < len(tentativi):
    guess = tentativi[indice]
    indice += 1
    if guess < segreto:
        print(f'  {guess}: troppo basso')
    elif guess > segreto:
        print(f'  {guess}: troppo alto')
    else:
        print(f'  {guess}: CORRETTO! in {indice} tentativi')
        break
else:
    print('Hai esaurito i tentativi')
````

## 10. Tuple

Una **tupla** è una sequenza **immutabile** e **ordinata** di elementi. Si usa quando i dati non devono essere modificati.

### Caratteristiche
- Immutabili: non si possono aggiungere, rimuovere o modificare elementi
- Più veloci delle liste per accesso e iterazione
- Hashable (se contengono solo tipi hashable): possono essere chiavi di dizionari o elementi di set
- Supportano unpacking (decompattazione)

### Quando usare tuple vs liste
| Tuple | Liste |
|-------|-------|
| Dati che non cambiano | Dati che cambiano |
| Coordinate, RGB, record | Elenchi, code, stack |
| Chiavi di dizionario | Non hashable |
| Return di funzioni multi-valore | Raccolta dinamica |


````python
# Creazione di tuple
vuota = ()
singola = (42,)             # la virgola è obbligatoria
punto = (3.0, 4.0)
colore_rgb = (255, 128, 0)  # arancione
mista = ('Alice', 17, True, 3.14)

print(f'Punto: {punto}')
print(f'Colore RGB: {colore_rgb}')
print(f'Lunghezza: {len(mista)}')

# Accesso per indice e slicing (come le liste)
print(f'Primo elemento: {colore_rgb[0]}')
print(f'Ultimo elemento: {colore_rgb[-1]}')
print(f'Slicing [1:]: {colore_rgb[1:]}')

# Immutabilità
try:
    colore_rgb[0] = 100    # genera TypeError
except TypeError as e:
    print(f'Errore (corretto!): {e}')

# Unpacking
x, y = punto
print(f'x={x}, y={y}')

nome, eta, *resto = mista
print(f'nome={nome}, eta={eta}, resto={resto}')

# Scambio di variabili con tuple unpacking
a, b = 10, 20
print(f'Prima: a={a}, b={b}')
a, b = b, a
print(f'Dopo scambio: a={a}, b={b}')

# Metodi delle tuple
dati = (3, 1, 4, 1, 5, 9, 2, 6, 5)
print(f'Conta il 5: {dati.count(5)}')
print(f'Indice del 9: {dati.index(9)}')
````

````python
# Named tuple – tuple con campi nominati
from collections import namedtuple

Studente = namedtuple('Studente', ['nome', 'eta', 'media'])

s1 = Studente('Alice', 17, 8.5)
s2 = Studente(nome='Bob', eta=18, media=7.2)

print(f'Nome: {s1.nome}, Media: {s1.media}')
print(f'Accesso per indice: {s1[0]}')

# NamedTuple con typing (Python 3.6+)
from typing import NamedTuple

class Punto3D(NamedTuple):
    x: float
    y: float
    z: float = 0.0  # valore default

p = Punto3D(1.0, 2.0)
print(f'Punto: {p}')
print(f'Coordinate: x={p.x}, y={p.y}, z={p.z}')
````

## 11. Liste

Una **lista** è una sequenza **mutabile** e **ordinata**. È la struttura dati più usata in Python.

### Metodi principali
| Metodo | Descrizione |
|--------|-------------|
| `append(x)` | Aggiunge x in fondo |
| `extend(iter)` | Aggiunge tutti gli elementi di iter |
| `insert(i, x)` | Inserisce x all'indice i |
| `remove(x)` | Rimuove la prima occorrenza di x |
| `pop([i])` | Rimuove e restituisce l'elemento in posizione i (default: -1) |
| `sort()` | Ordina in-place |
| `sorted()` | Restituisce una nuova lista ordinata |
| `reverse()` | Inverte in-place |
| `index(x)` | Indice della prima occorrenza |
| `count(x)` | Numero di occorrenze |
| `copy()` | Copia superficiale |
| `clear()` | Svuota la lista |


````python
# Creazione
vuota = []
numeri = [1, 2, 3, 4, 5]
mista = ['ciao', 42, True, 3.14, None]

# Slicing [start:stop:step]
seq = list(range(10))  # [0,1,2,...,9]
print(f'Lista completa: {seq}')
print(f'Primi 3: {seq[:3]}')
print(f'Ultimi 3: {seq[-3:]}')
print(f'Dal 2 al 7 (step 2): {seq[2:8:2]}')
print(f'Invertita: {seq[::-1]}')

# Metodi
frutti = ['mela', 'banana']
frutti.append('ciliegia')
frutti.extend(['dattero', 'fico'])
frutti.insert(1, 'albicocca')
print(f'Dopo append/extend/insert: {frutti}')

frutti.remove('banana')
ultimo = frutti.pop()
print(f'Dopo remove e pop: {frutti}, rimosso: {ultimo}')

# Ordinamento
voti = [7.5, 9.0, 6.5, 8.0, 9.5, 7.0]
print(f'Originale: {voti}')
print(f'sorted(): {sorted(voti)}')
print(f'sorted(reverse=True): {sorted(voti, reverse=True)}')

nomi = ['Carla', 'alice', 'Bob']
print(f'Sort case-insensitive: {sorted(nomi, key=str.lower)}')
````

## 12. Dizionari

### Metodi principali
| Metodo | Descrizione |

| `dict.get(key, default)` | Accede senza eccezione |

````python
# Creazione
vuoto = {}
studente = {
    'nome': 'Alice',
    'eta': 17,
    'classe': '4A',
    'voti': [8, 7, 9, 8.5]
}

# Accesso
print(f'Nome: {studente["nome"]}')
print(f'Voti: {studente["voti"]}')

# get() – sicuro
print(f'Media: {studente.get("media", "N/A")}')

# Modifica e aggiunta
studente['eta'] = 18
studente['media'] = sum(studente['voti']) / len(studente['voti'])
print(f'Dopo modifica: eta={studente["eta"]}, media={studente["media"]}')

# Iterazione
print('\nChiavi:')
for chiave in studente.keys():
    print(f'  {chiave}')

print('\nCoppie chiave-valore:')
for k, v in studente.items():
    print(f'  {k}: {v}')

# pop
eta_rimossa = studente.pop('eta')
print(f'\nEtà rimossa: {eta_rimossa}')
print(f'Dict ora: {studente}')
````

## 14. Funzioni

Una **funzione** è un blocco di codice riutilizzabile definito con `def`.

### Tipi di parametri
| Tipo | Sintassi | Descrizione |
|------|----------|-------------|
| Posizionale | `def f(a, b)` | Passati per posizione |
| Default | `def f(a, b=10)` | Valore predefinito se omesso |
| *args | `def f(*args)` | Numero variabile di posizionali → tuple |
| **kwargs | `def f(**kwargs)` | Numero variabile di keyword → dict |
| Solo keyword | `def f(*, k)` | Devono essere passati per nome |

### Funzioni lambda
Funzioni anonime su una riga: `lambda parametri: espressione`

### Ricorsione
Una funzione che chiama se stessa. Necessita di un **caso base** per terminare.


````python
# Definizione e chiamata base
def saluta(nome, messaggio='Ciao'):
    """Saluta una persona con un messaggio."""
    return f'{messaggio}, {nome}!'

print(saluta('Alice'))
print(saluta('Bob', 'Buongiorno'))
print(saluta(messaggio='Salve', nome='Carla'))  # keyword args

# *args – argomenti variabili posizionali
def somma(*numeri):
    """Somma un numero arbitrario di valori."""
    return sum(numeri)

print(f'\nSomma(1,2,3) = {somma(1, 2, 3)}')
print(f'Somma(1..10) = {somma(*range(1, 11))}')

# **kwargs – argomenti keyword arbitrari
def crea_profilo(**kwargs):
    """Crea un profilo da coppie chiave=valore."""
    return {k: v for k, v in kwargs.items()}

profilo = crea_profilo(nome='Alice', eta=17, classe='4A')
print(f'\nProfilo: {profilo}')

# Funzione con tutti i tipi di parametro
def completa(a, b, *args, sep=', ', **kwargs):
    print(f'a={a}, b={b}, args={args}, sep={sep!r}, kwargs={kwargs}')

completa(1, 2, 3, 4, 5, sep=' | ', x=10, y=20)
````

````python
# Lambda – funzioni anonime
quadrato = lambda x: x**2
somma = lambda a, b: a + b
print(f'quadrato(7) = {quadrato(7)}')
print(f'somma(3, 5) = {somma(3, 5)}')

# Lambda con sorted, map, filter
studenti = [
    {'nome': 'Alice', 'media': 8.5},
    {'nome': 'Bob',   'media': 7.0},
    {'nome': 'Carla', 'media': 9.2},
]

# Ordina per media decrescente
per_media = sorted(studenti, key=lambda s: s['media'], reverse=True)
for s in per_media:
    print(f'  {s["nome"]}: {s["media"]}')

# map e filter
numeri = list(range(1, 11))
quadrati = list(map(lambda x: x**2, numeri))
dispari = list(filter(lambda x: x % 2 != 0, numeri))
print(f'\nQuadrati: {quadrati}')
print(f'Dispari:  {dispari}')

# Ricorsione
def fattoriale(n):
    """n! con ricorsione."""
    if n <= 1:           # caso base
        return 1
    return n * fattoriale(n - 1)  # caso ricorsivo

def fibonacci(n):
    """Fibonacci con memoization."""
    memo = {}
    def _fib(k):
        if k in memo:
            return memo[k]
        if k <= 1:
            return k
        memo[k] = _fib(k-1) + _fib(k-2)
        return memo[k]
    return _fib(n)

for i in range(8):
    print(f'  {i}! = {fattoriale(i)}')
print(f'\nFibonacci 0-9: {[fibonacci(i) for i in range(10)]}')
````

## 15. Gestione delle Eccezioni

Le eccezioni sono eventi che interrompono il normale flusso del programma. Python permette di gestirle con `try/except`.

### Struttura completa
```python
try:
    # codice che potrebbe lanciare eccezione
except TipoEccezione as e:
    # gestione dell'errore
except (Tipo1, Tipo2) as e:
    # gestione multipla
else:
    # eseguito SOLO se nessuna eccezione
finally:
    # eseguito SEMPRE (cleanup)
```

### Eccezioni comuni
| Eccezione | Causa |
|-----------|-------|
| `ValueError` | Valore non valido |
| `TypeError` | Tipo non compatibile |
| `IndexError` | Indice fuori range |
| `KeyError` | Chiave assente in dict |
| `FileNotFoundError` | File non trovato |
| `ZeroDivisionError` | Divisione per zero |
| `AttributeError` | Attributo inesistente |
| `ImportError` | Modulo non trovabile |


````python
# try/except base
def dividi(a, b):
    try:
        risultato = a / b
    except ZeroDivisionError:
        print(f'  Errore: divisione per zero ({a} / {b})')
        return None
    except TypeError as e:
        print(f'  Tipo non valido: {e}')
        return None
    else:
        print(f'  {a} / {b} = {risultato}')
        return risultato
    finally:
        print(f'  [finally eseguito sempre]')

dividi(10, 2)
print()
dividi(10, 0)
print()
dividi('ciao', 3)

# Cattura multipla
dati = ['10', '3.14', 'ciao', None, '42']
print('\nConversione dati:')
for d in dati:
    try:
        print(f'  int({d!r}) = {int(d)}')
    except (ValueError, TypeError) as e:
        print(f'  int({d!r}) → Errore: {e}')
````

````python
# Eccezioni personalizzate
class VotoNonValido(ValueError):
    """Eccezione per voti fuori dal range valido."""
    def __init__(self, voto, minimo=0, massimo=10):
        self.voto = voto
        self.minimo = minimo
        self.massimo = massimo
        super().__init__(
            f'Voto {voto} non valido: deve essere tra {minimo} e {massimo}'
        )

class StudenteNonTrovato(KeyError):
    """Eccezione per studente inesistente nel registro."""
    pass

def registra_voto(registro, nome, voto):
    if not (0 <= voto <= 10):
        raise VotoNonValido(voto)
    if nome not in registro:
        raise StudenteNonTrovato(f'Studente {nome!r} non trovato')
    registro[nome].append(voto)

registro = {'Alice': [], 'Bob': []}

operazioni = [
    ('Alice', 8),
    ('Bob', 11),    # voto non valido
    ('Carla', 7),   # studente non trovato
    ('Alice', 9),
]

for nome, voto in operazioni:
    try:
        registra_voto(registro, nome, voto)
        print(f'  OK: {nome} → {voto}')
    except VotoNonValido as e:
        print(f'  VotoNonValido: {e}')
    except StudenteNonTrovato as e:
        print(f'  StudenteNonTrovato: {e}')

print(f'\nRegistro finale: {registro}')
````

## 18. Classi e Metodi OOP

La **Programmazione Orientata agli Oggetti** (OOP) organizza il codice intorno a **oggetti** che combinano dati (attributi) e comportamenti (metodi).

### Concetti fondamentali
| Concetto | Descrizione |
|----------|-------------|
| **Classe** | Schema/template per creare oggetti |
| **Oggetto/Istanza** | Concretizzazione di una classe |
| **Attributo** | Dato appartenente all'oggetto |
| **Metodo** | Funzione appartenente alla classe |
| **`__init__`** | Costruttore: inizializza l'oggetto |
| **`self`** | Riferimento all'istanza corrente |

### Dunder methods (magic methods)
| Metodo | Quando viene chiamato |
|--------|----------------------|
| `__init__` | Costruzione oggetto |
| `__str__` | `str(obj)`, `print(obj)` |
| `__repr__` | Rappresentazione debug |
| `__len__` | `len(obj)` |
| `__eq__` | `obj1 == obj2` |
| `__lt__` | `obj1 < obj2` |
| `__add__` | `obj1 + obj2` |


````python
class Studente:
    """Rappresenta uno studente dell'istituto."""

    # Attributo di classe (condiviso da tutte le istanze)
    scuola = 'ITIS Galileo'
    _contatore = 0

    def __init__(self, nome, cognome, classe):
        """Costruttore."""
        self._nome = nome           # attributo 'protetto'
        self._cognome = cognome
        self._classe = classe
        self._voti = []
        Studente._contatore += 1
        self._id = Studente._contatore

    # Property: accesso controllato agli attributi
    @property
    def nome_completo(self):
        return f'{self._nome} {self._cognome}'

    @property
    def media(self):
        if not self._voti:
            return 0.0
        return sum(self._voti) / len(self._voti)

    def aggiungi_voto(self, voto):
        """Aggiunge un voto (deve essere tra 0 e 10)."""
        if not 0 <= voto <= 10:
            raise ValueError(f'Voto {voto} non valido')
        self._voti.append(voto)
        return self

    # Classmethod: lavora sulla classe, non sull'istanza
    @classmethod
    def contatore(cls):
        return cls._contatore

    # Staticmethod: funzione utility senza self o cls
    @staticmethod
    def valida_voto(voto):
        return 0 <= voto <= 10

    # Dunder methods
    def __str__(self):
        return f'{self.nome_completo} ({self._classe}) - Media: {self.media:.2f}'

    def __repr__(self):
        return f'Studente(id={self._id}, nome={self._nome!r}, classe={self._classe!r})'

    def __len__(self):
        return len(self._voti)

    def __eq__(self, altro):
        return self._id == altro._id


# Utilizzo
s1 = Studente('Alice', 'Rossi', '4A')
s1.aggiungi_voto(8).aggiungi_voto(9).aggiungi_voto(7.5)  # method chaining

s2 = Studente('Bob', 'Verdi', '4B')
s2.aggiungi_voto(6).aggiungi_voto(7)

print(s1)            # chiama __str__
print(repr(s2))      # chiama __repr__
print(f'Voti di Alice: {len(s1)}')  # chiama __len__
print(f'Studenti totali: {Studente.contatore()}')
print(f'Voto 11 valido? {Studente.valida_voto(11)}')
print(f's1 == s2: {s1 == s2}')
````

## 19. Ereditarietà e Polimorfismo

L'**ereditarietà** permette a una classe di ereditare attributi e metodi da una classe genitore.
Il **polimorfismo** permette a oggetti di classi diverse di rispondere allo stesso metodo in modo differente.

### Keyword chiave
- `class Figlia(Genitore):` – definisce la genitrice
- `super()` – riferimento alla classe genitore
- `isinstance(obj, Classe)` – verifica il tipo
- `issubclass(Figlia, Genitore)` – verifica la gerarchia

### Tipi di ereditarietà
- **Singola**: una sola classe genitore
- **Multipla**: più classi genitrici (Python supporta MRO – Method Resolution Order)
- **Mixin**: classe progettata per aggiungere funzionalità


````python
class Persona:
    """Classe base."""
    def __init__(self, nome, eta):
        self.nome = nome
        self.eta = eta

    def presentati(self):
        return f'Sono {self.nome}, ho {self.eta} anni'

    def __str__(self):
        return f'Persona({self.nome}, {self.eta})'


class Insegnante(Persona):
    """Eredita da Persona."""
    def __init__(self, nome, eta, materia):
        super().__init__(nome, eta)  # chiama __init__ del genitore
        self.materia = materia

    def presentati(self):  # override del metodo
        base = super().presentati()
        return f'{base}, insegno {self.materia}'

    def __str__(self):
        return f'Insegnante({self.nome}, {self.materia})'


class StudenteIT(Persona):
    """Studente istituto tecnico."""
    def __init__(self, nome, eta, classe, specializzazione):
        super().__init__(nome, eta)
        self.classe = classe
        self.specializzazione = specializzazione

    def presentati(self):
        base = super().presentati()
        return f'{base}, classe {self.classe} ({self.specializzazione})'


# Istanze
persone = [
    Persona('Mario', 45),
    Insegnante('Prof. Bianchi', 52, 'Informatica'),
    StudenteIT('Alice', 17, '4A', 'Informatica'),
    StudenteIT('Bob', 18, '4B', 'Elettronica'),
]

# Polimorfismo: stesso metodo, comportamento diverso
print('=== Polimorfismo ===")
for p in persone:
    print(f'  {p.presentati()}')

# isinstance e issubclass
print('\n=== isinstance ===")
alice = persone[2]
print(f'Alice è Persona? {isinstance(alice, Persona)}')
print(f'Alice è StudenteIT? {isinstance(alice, StudenteIT)}')
print(f'Alice è Insegnante? {isinstance(alice, Insegnante)}')
print(f'StudenteIT è subclass di Persona? {issubclass(StudenteIT, Persona)}')

# MRO – Method Resolution Order
print(f'\nMRO di StudenteIT: {[c.__name__ for c in StudenteIT.__mro__]}')
````

## 20. Threads (Programmazione Concorrente)

Un **thread** è un flusso di esecuzione indipendente all'interno dello stesso processo. Python usa il modulo `threading`.

> **GIL (Global Interpreter Lock)**: CPython permette l'esecuzione di un solo thread Python alla volta. I thread sono utili per I/O-bound tasks (rete, file). Per CPU-bound tasks si usa `multiprocessing`.

### Concetti fondamentali
| Concetto | Descrizione |
|----------|-------------|
| `Thread` | Oggetto che rappresenta un thread |
| `start()` | Avvia il thread |
| `join()` | Attende la terminazione |
| `Lock` | Mutex per sezioni critiche |
| `Queue` | Coda thread-safe per producer-consumer |
| `Event` | Segnale tra thread |
| `daemon` | Thread che termina col processo principale |

### Race condition
Si verifica quando due thread accedono concorrentemente a una risorsa condivisa senza sincronizzazione, producendo risultati non deterministici.


````python
import threading
import time

# Thread base
def worker(nome, secondi):
    print(f'[{nome}] Inizio (sleep {secondi}s)')
    time.sleep(secondi)
    print(f'[{nome}] Fine')

print('=== Thread Base ===")
t1 = threading.Thread(target=worker, args=('Thread-1', 0.2))
t2 = threading.Thread(target=worker, args=('Thread-2', 0.1))
t3 = threading.Thread(target=worker, args=('Thread-3', 0.15))

avvio = time.time()
t1.start(); t2.start(); t3.start()
t1.join(); t2.join(); t3.join()
print(f'Completato in {time.time()-avvio:.3f}s (sequenziale sarebbe ~0.45s)')

# Race condition e Lock
print('\n=== Lock (Mutex) ===")
contatore = 0
lock = threading.Lock()

def incrementa(volte, con_lock):
    global contatore
    for _ in range(volte):
        if con_lock:
            with lock:
                contatore += 1
        else:
            contatore += 1  # RACE CONDITION!

# Con lock
contatore = 0
threads = [threading.Thread(target=incrementa, args=(1000, True)) for _ in range(5)]
for t in threads: t.start()
for t in threads: t.join()
print(f'Con Lock: contatore = {contatore} (atteso: 5000)')
````

````python
import threading
import queue
import time
import random

# Producer-Consumer con Queue (thread-safe)
print('=== Producer-Consumer ===")

coda = queue.Queue(maxsize=5)  # buffer limitato
risultati = []
SENTINEL = object()            # segnale di stop

def producer(n_elementi):
    """Produce n_elementi e li mette in coda."""
    for i in range(n_elementi):
        elemento = f'item-{i}'
        coda.put(elemento)     # blocca se coda piena
        print(f'  [Producer] prodotto: {elemento} (coda: {coda.qsize()})')
        time.sleep(random.uniform(0.02, 0.08))
    coda.put(SENTINEL)         # segnale di fine
    print(f'  [Producer] terminato')

def consumer(nome):
    """Consuma dalla coda finché non riceve SENTINEL."""
    while True:
        elemento = coda.get()  # blocca se coda vuota
        if elemento is SENTINEL:
            coda.put(SENTINEL) # rimetti per altri consumer
            break
        risultati.append(elemento)
        print(f'  [{nome}] consumato: {elemento}')
        time.sleep(random.uniform(0.05, 0.1))
        coda.task_done()
    print(f'  [{nome}] terminato')

p = threading.Thread(target=producer, args=(6,))
c1 = threading.Thread(target=consumer, args=('Consumer-1',))
c2 = threading.Thread(target=consumer, args=('Consumer-2',))

p.start(); c1.start(); c2.start()
p.join(); c1.join(); c2.join()

print(f'\nElementi consumati: {len(risultati)} → {sorted(risultati)}')
````

## 21. Socket TCP

I **socket** sono l'astrazione di base per la comunicazione di rete. TCP (Transmission Control Protocol) garantisce:
- **Connessione orientata** (3-way handshake)
- **Affidabilità** (ritrasmissione in caso di perdita)
- **Ordine** (i dati arrivano nell'ordine inviato)

### Flusso Server TCP
```
socket() → bind() → listen() → accept() → recv/send() → close()
```

### Flusso Client TCP
```
socket() → connect() → send/recv() → close()
```

### Parametri socket
| Parametro | Valore | Descrizione |
|-----------|--------|-------------|
| `AF_INET` | `socket.AF_INET` | IPv4 |
| `SOCK_STREAM` | `socket.SOCK_STREAM` | TCP |
| `SOCK_DGRAM` | `socket.SOCK_DGRAM` | UDP |


````python
import socket
import threading
import time

# Demo TCP: server e client nello stesso notebook
HOST = '127.0.0.1'
PORT_TCP = 55001
log = []  # raccoglie i messaggi per visualizzarli

def tcp_server():
    """Server TCP echo: restituisce il messaggio in maiuscolo."""
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as srv:
        srv.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
        srv.bind((HOST, PORT_TCP))
        srv.listen(3)
        log.append(f'[SERVER] In ascolto su {HOST}:{PORT_TCP}')

        conn, addr = srv.accept()
        with conn:
            log.append(f'[SERVER] Connessione da {addr}')
            while True:
                dati = conn.recv(1024)
                if not dati:
                    break
                messaggio = dati.decode('utf-8')
                log.append(f'[SERVER] Ricevuto: {messaggio!r}')
                risposta = messaggio.upper()
                conn.sendall(risposta.encode('utf-8'))
                log.append(f'[SERVER] Inviato: {risposta!r}')
        log.append('[SERVER] Connessione chiusa')

def tcp_client(messaggi):
    """Client TCP che invia una lista di messaggi."""
    time.sleep(0.1)  # attende che il server sia pronto
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as cli:
        cli.connect((HOST, PORT_TCP))
        log.append(f'[CLIENT] Connesso a {HOST}:{PORT_TCP}')
        for msg in messaggi:
            cli.sendall(msg.encode('utf-8'))
            log.append(f'[CLIENT] Inviato: {msg!r}')
            risposta = cli.recv(1024).decode('utf-8')
            log.append(f'[CLIENT] Ricevuto: {risposta!r}')
            time.sleep(0.05)
    log.append('[CLIENT] Disconnesso')

# Avvia server e client in thread separati
messaggi_test = ['ciao server', 'python e' ' socket', 'buona fortuna']

ts = threading.Thread(target=tcp_server, daemon=True)
tc = threading.Thread(target=tcp_client, args=(messaggi_test,))

ts.start()
tc.start()
tc.join(timeout=5)
time.sleep(0.2)

print('\n'.join(log))
````

## 22. Socket UDP

UDP (User Datagram Protocol) è **senza connessione** e **non affidabile**: i pacchetti possono perdersi o arrivare fuori ordine. È più veloce di TCP.

### Quando usare UDP
- Streaming audio/video (meglio perdere un frame che ritardare)
- Gaming online (latenza bassa è prioritaria)
- DNS, DHCP, NTP
- Broadcast e multicast

### Differenze TCP vs UDP
| Caratteristica | TCP | UDP |
|---------------|-----|-----|
| Connessione | Sì | No |
| Affidabilità | Sì | No |
| Ordine | Garantito | Non garantito |
| Velocità | Più lento | Più veloce |
| Overhead | Alto | Basso |
| Uso tipico | HTTP, FTP, SSH | DNS, streaming, gaming |


````python
import socket
import threading
import time

HOST = '127.0.0.1'
PORT_UDP = 55002
log_udp = []

def udp_server(n_messaggi):
    """Server UDP: riceve n_messaggi e risponde con ACK."""
    with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as srv:
        srv.bind((HOST, PORT_UDP))
        log_udp.append(f'[UDP SERVER] Bind su {HOST}:{PORT_UDP}')
        for _ in range(n_messaggi):
            dati, addr = srv.recvfrom(1024)
            msg = dati.decode('utf-8')
            log_udp.append(f'[UDP SERVER] Da {addr}: {msg!r}')
            ack = f'ACK:{msg.upper()}'
            srv.sendto(ack.encode('utf-8'), addr)
        log_udp.append('[UDP SERVER] Fine')

def udp_client(messaggi):
    """Client UDP: invia messaggi e riceve ACK."""
    time.sleep(0.1)
    with socket.socket(socket.AF_INET, socket.SOCK_DGRAM) as cli:
        cli.settimeout(2.0)  # timeout per recvfrom
        for msg in messaggi:
            cli.sendto(msg.encode('utf-8'), (HOST, PORT_UDP))
            log_udp.append(f'[UDP CLIENT] Inviato: {msg!r}')
            try:
                ack, _ = cli.recvfrom(1024)
                log_udp.append(f'[UDP CLIENT] ACK: {ack.decode()}')
            except socket.timeout:
                log_udp.append('[UDP CLIENT] Timeout!')
            time.sleep(0.05)
    log_udp.append('[UDP CLIENT] Fine')

messaggi_udp = ['ping', 'hello udp', 'terzo pacchetto']

ts = threading.Thread(target=udp_server, args=(len(messaggi_udp),), daemon=True)
tc = threading.Thread(target=udp_client, args=(messaggi_udp,))

ts.start()
tc.start()
tc.join(timeout=5)
time.sleep(0.2)

print('\n'.join(log_udp))
````
